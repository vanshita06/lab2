```python
 English_Stopwords = list(nltk.corpus.stopwords.words('english'))
English_Stopwords
```

```python
l=['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves']
print(l)
```

```python
for i in l:
    if i in English_Stopwords:
        English_Stopwords.remove(i)
English_Stopwords
```
